```python
import numpy as np 
import pandas as pd
import os

files = os.listdir()
size = len(files)
print("Total samples:",size)
print(files[0])
```

```python
import cv2
images = []
ages = []
genders = []
for file in files:
    #image = cv2.imread(file,0)
    #image = cv2.resize(image,dsize=(227,227))
    #image = image.reshape((image.shape[0],image.shape[1],1))
    #images.append(image)
    split_var = file.split('_')
    ages.append(split_var[0])
    genders.append(int(split_var[1]))
```

```python
processed_age = []
print(ages)
for i in ages:
  i=int(i)
  if i<=3:
    processed_age.append('0-2')
  elif i <= 7:
    processed_age.append('4-6')
  elif i <= 13:
    processed_age.append('8-13')
  elif i <= 22:
    processed_age.append('15-20')
  elif i <= 34:
    processed_age.append('25-32')
  elif i <= 45:
    processed_age.append('38-43')
  elif i <= 59:
    processed_age.append('48-53')
  else:
    processed_age.append('60+')

print(processed_age)

```

```python
print(genders)
```

```python
processed_data = list(zip(files, processed_age, genders))
len(processed_data)
```

```python
output_file = open("UTK_processed_data.txt", 'w')
for i in processed_data:
  output_file.write(str(i)+'\n')
output_file.close()
```

```python
for file in files:
    image = cv2.imread(file,0)
    image = cv2.resize(image,dsize=(227,227))
    image = image.reshape((image.shape[0],image.shape[1],1))
    images.append(image)
```

```python
images = pd.Series(list(images), name = 'Images')
ages = pd.Series(list(ages), name = 'Ages')
genders = pd.Series(list(genders), name = 'Genders')

df = pd.concat([images, ages, genders], axis=1)
df
```
